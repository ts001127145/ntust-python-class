### pandas使用
* **創造DataFrame**：pd.DataFrame(list)
* **讀取資料**：pd.read_csv(...)/pd.read.json(...)
* **選擇特定欄位資料**：df = pd.read_csv(...); pd["column_name"]
* **造新欄位**：df["column name"] = df["old_column"]
(ex: df["重量磅" = df ["重量(磅)"]=df["重量"/1000])
* **過濾資料(將條件放入)**：df(condition)
(ex: df[df["國家"] == "日本"]，裡面的df意思是從DataFrame裡面判斷國家是日本的欄位，外面的df意思是再把符合條件的項目丟回原先的DataFrame並顯示出符合條件的欄位)

### 小試身手

#### 試撰寫一個class為Student
#### 靜態屬性包含：standard = 70
#### 屬性包含：name(str)，studNo(int), grades(list)
#### 函式包含：
    * avg：回傳grades list的平均值
    * add(grade)：新增成績到grades list中，並顯示grades狀態
    * failedCount()：回傳不及格(<standard)總數

```python
class Student:
    standard = 70
    def __init__(self, name, studNo, grades, standard):
        self.name = name
        self.studNo = studNo
        self.grades = grades
        self.standard = standard
    def avg(self):
        return sum(self.grades)/3
    
    
    
    def failedCount(self):
        result = []  # 先準備一個空陣列
        for score in self.grades:
            if score < Student.standard:
                result.append(score)
        return len(result)
```

```python
# 試跑資料
s1 = Student("Kevin", 1001, [90,50,35])
s2 = Student("Jenny", 1002, [72,84,99])

#s1.avg
#s2.avg

```
